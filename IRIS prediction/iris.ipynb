```python
#Using a Decision Tree model to predict the class of the flowers
from sklearn.model_selection import train_test_split #this imports the train_test_split function
from sklearn.tree import DecisionTreeClassifier#Importing the Decision tree model
from sklearn.metrics import accuracy_score# imports accuracy score function
iris_data=pd.read_csv("Iris.csv")#read the iris dataset
print(iris_data.head())#show the first four rows of the dataset  
```

```python
print(iris_data.describe())#Describe what the data does 
```

```python
X=iris_data.drop("Species", axis=1)#drop the species part of the data
y=iris_data["Species"]#Use the species as the output variable 
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)#this splits the data
model=DecisionTreeClassifier(random_state=42)#using the random_state for shuffling data
model.fit(X_train,y_train)#This line makes the model learn the data
unseen=model.predict(X_test)#This line shows that the model should predict the unseen data
accuracy=accuracy_score(y_test,unseen)#predicting the accuracy of the model using test data
print(f"Accuracy is,{accuracy}")#printing the accuracy of the model


```

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
import pandas as pd

iris=pd.read_csv("Iris.csv")


# Drop 'Id' and 'Species' from the features (X)
X = iris.drop(["Id","Species"], axis=1)

# Use 'Species' as the label (y)
y = iris["Species"]

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

# Train model
model = DecisionTreeClassifier(random_state=42)

model.fit(X_train, y_train)

# Confirm number of features
print("Model expects:", model.n_features_in_)  # Should be 4

# Now collect user input
Sepal_length = float(input("Enter your sepal length: "))
Sepal_width = float(input("Enter your sepal width: "))
Petal_length = float(input("Enter your petal length: "))
Petal_width = float(input("Enter your petal width: "))

# Predict
features = [[Sepal_length, Sepal_width, Petal_length, Petal_width]]
prediction = model.predict(features)
print(f'The predicted flower is: {prediction}')
```
